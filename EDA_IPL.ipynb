Which team has the highest winning percentage?


```python
import pandas as pd

total_matches_played = pd.concat([df2['team1'], df2['team2']]).value_counts()
actual_winners = df2[df2['winner'] != 'no_winner']['winner']
matches_won = actual_winners.value_counts()
combined_stats = pd.DataFrame({'played': total_matches_played, 'won': matches_won}).fillna(0)
winning_percentage = (combined_stats['won'] / combined_stats['played'] * 100).fillna(0)
highest_winning_team = winning_percentage.idxmax()
highest_percentage = winning_percentage.max()

print(f"The team with the highest winning percentage is '{highest_winning_team}' with {highest_percentage:.2f}% wins.")
```
